## Popularity based Recommender System

```python
book_ratings_with_name = ratings_df.merge(books_df, on='ISBN')
book_ratings_with_name.head()
```

```python
number_of_ratings = book_ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
number_of_ratings
#reset_index() - converts the pandas series obtained fromt count into a dataframe
```

```python
number_of_ratings[number_of_ratings['Book-Title']=='Flesh Tones: A Novel']
```

```python
number_of_ratings.rename(columns={'Book-Rating':'Number of ratings'}, inplace=True)
```

```python
number_of_ratings
```

```python
avg_of_ratings = book_ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_of_ratings.rename(columns={'Book-Rating':'Avg of ratings'}, inplace=True)
avg_of_ratings
```

```python
popularity_df = number_of_ratings.merge(avg_of_ratings, on='Book-Title')
```

```python
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['Number of ratings']>=250].sort_values('Avg of ratings', ascending=False).head(50)
```

```python
popularity_df
```

```python
top50 = popularity_df.merge(books_df, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Publisher', 'Image-URL-S', 'Number of ratings', 'Avg of ratings']]
```

## Collaborative based Recommender System

```python
#getting users who have given at least 150 ratings
temp = book_ratings_with_name.groupby('User-ID').count()['Book-Rating']>=150
useful_users = temp[temp].index #boolean indexing. consists of index of users who have given more than 150 reviews
```

```python
useeful_ratings = book_ratings_with_name[book_ratings_with_name['User-ID'].isin(useful_users)]
```

```python
t = useeful_ratings.groupby('Book-Title').count()['Book-Rating']>=60
famous_books = t[t].index
```

```python
final_ratings = useeful_ratings[useeful_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
final_df = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
final_df.fillna(0, inplace=True)
```

```python
final_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(final_df)
similarity
```

```python
sorted(list(enumerate(similarity[0])), key=lambda x: x[1], reverse=True)
```

```python
def recommend(book_name):
    index = np.where(final_df.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity[index])), key=lambda x: x[1], reverse=True)[1:6]
    
    data = []
    for book in similar_books:
        item = []
        temp_df = books_df[books_df["Book-Title"]==final_df.index[book[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    
    return data
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
